# 4-Class motor imagery (Pacman)

Implemented through extensions systems in the [PacmanMotorImagery](../../../../_modules/bci_framework.default_extensions.Class_motor_imagery_Pacman.main.rst) module.

<img src='paradigm.png'></img>

This cue-based BCI paradigm consisted of up to four different motor imagery tasks, represented by the classic Pacman character displacements and separated with an asynchronous break. This paradigm implements a clean interface, unlike the game-like design used in <cite data-footcite="perdikis2014subject">perdikis2014subject</cite>.


## Configuration options

<img src='../../../images/extensions/4class_mi_pacman.png'></img>

**Cues:**  
Is possible to select up to 4 classes, represented by one arrow. All 4 classes are active by default.

**Trials per class:**  
How many trials per class will contain the run. Between 1 and 100, 10 by default.

**Stimulus onset asynchronously:**  
To prevent fixed time between trials. A range between 1000 and 3000 ms, (2000-3000) by default. This means that for each trial a random number (with uniform distribution) is generated between this range and used as inter stimulus interval.
