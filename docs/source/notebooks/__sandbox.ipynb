```python
# ----------------------------------------------------------------------
def _get_factor_near_to(x, n=1000):
    """Get the integer number factor of `x` nearest to `n`.

    This factor is used to fast resampling.

    Parameters
    ----------
    x
        Samples.
    n
        Near factor

    Returns
    -------
    int
        Factor.
    """
    a = np.array([(x) / np.arange(max(1, (x // n) - 10), (x // n) + 10)])[0]
    a[a % 1 != 0] = 0
    return int(a[np.argmin(np.abs(a - n))])
```
